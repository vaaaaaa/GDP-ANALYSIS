```python
df_china=df[df['Country Name']=='Hong Kong SAR, China']
df_china.loc[:, 'GDP_change'] = df_china['Value'].pct_change() * 100
df_china.loc[0, 'GDP_change'] = 0
df_china
```

```python
final_data=[]
for country_name in df['Country Name'].unique():
  df_temp=df[df['Country Name']==country_name]
  data=df_temp.values

  gdp_change=[0]

  for i in range(1,len(data)):
    prev=data[i-1][3]
    curr=data[i][3]
    gdp_change.append(round((curr-prev)/prev*100,2))
  df_temp=df_temp.copy()
  df_temp=df_temp.assign(GDP_change=gdp_change)
  final_data.append(df_temp)


```

```python
df=pd.concat(final_data)
```

## Multiple Countries Comparision

```python
list=['China','India','World','United States']
data=[]
for i in list:
  df_compare=df[df['Country Name']==i]
  data.append(df_compare)
  df_compare=pd.concat(data,axis=0)

fig=px.line(df_compare,x='Year',y='Value',title='|'.join(list)+'GDP',color='Country Name')
fig.show()
pyo.plot(fig,filename=f'{i} GDP.html')



```

```python
def GDP_Comparision(list,isopen):
  data=[]
  for i in list:
    df_compare=df[df['Country Name']==i]
    data.append(df_compare)
    df_compare=pd.concat(data,axis=0)
  fig=px.line(df_compare,x='Year',y='Value',title='|'.join(list)+'GDP',color='Country Name')
  fig.show()
  pyo.plot(fig,filename=f'{i} GDP.html',auto_open=isopen)
```

```python
GDP_Comparision(['China','India','United States'],True)
```
